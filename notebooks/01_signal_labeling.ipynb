# 01 信号标注

这个 notebook 用来把主观案例整理为标准样本、边界样本和失败样本。

建议流程：
1. 先加载单只股票历史数据。
2. 运行规则引擎查看系统标注。
3. 手工判断它是否符合你的原始定义。
4. 回写到 `data/samples/*.csv`。

```python
from pathlib import Path

import pandas as pd

from app.data.ingest import DataIngestConfig, fetch_a_share_history
from app.report.charting import plot_signal_context
from app.strategy.rules import scan_signal_history

PROJECT_ROOT = Path.cwd().parent
history = fetch_a_share_history("000001", DataIngestConfig(cache_dir=PROJECT_ROOT / "data" / "cache"))
signals = scan_signal_history(history, symbol="000001", enabled_signals=["2b_structure", "double_breakout"], include_invalid=False)
signal_frame = pd.DataFrame([signal.to_dict() for signal in signals])
signal_frame.tail(10)
```

```python
# 选一条历史信号导出图，便于人工判断它是否属于标准样本或边界样本
selected = signals[-1]
chart_path = PROJECT_ROOT / "reports" / "notebook_charts" / f"{selected.symbol}_{selected.signal_date.isoformat()}_{selected.signal_type}.png"
plot_signal_context(history, selected, chart_path)
```
